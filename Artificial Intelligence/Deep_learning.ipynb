```python
import torch
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
# data
X = torch.tensor([50.,80.,120.])
y = torch.tensor([25.,40.,60.])
```

```python
# normalize
X_mean , X_std = X.mean(),X.std()
y_norm = (X - X_mean) / X_std
```

```python
# model - weight and bias

w = torch.tensor([0.0],requires_grad=True)
b = torch.tensor([0.0],requires_grad=True)

optimizer = optim.SGD([w,b],lr=0.001)  # gradient descent
```

```python
losses = []

for epoch in range(1200):
  # forword
  y_pred = w * y_norm + b

  # loass
  loss = ((y_pred - y)**2).mean()
  losses.append(loss.item())

  # backward
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  if epoch %100 == 0:
    print(f"Epoch {epoch:4d} | loss {loss.item():8.4f} | w {w.item():6.3f} | b {b.item():5.2f}")
```

```python
# final output
print(f"price  = {w.item()/X_std.item():.4f}")
```
